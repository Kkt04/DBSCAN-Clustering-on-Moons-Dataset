```python
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
X, y = make_moons(n_samples= 350, noise = 0.05, random_state= 42)
```

```python
X.shape
```

```python
#visualize the scatter plot
sns.scatterplot(x=X[:,0], y=X[:,1])
```

```python
#try to scale the dataset using the standardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled[:5]
```

```python
# lets try to import DBSCAN from sklearn
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# take eps = 0.3 (just as an example)
dbscan = DBSCAN (eps = 0.3, min_samples = 10)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
#visualize the scatter plt for DBSCAN
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=dbscan.labels_)
```
